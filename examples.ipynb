```python
import numpy as np
import matplotlib.pyplot as plt
from potential1D import Potential1D
```

# Basis

$$
\begin{align}
f_n(\omega) =
\left\{ 
    \begin{matrix}
    \sqrt{\frac{1}{T}}, & n=0 \\
    \sqrt{\frac{2}{T}}\cos(n \omega), & 0< n \le \frac{N+1}{2} \\
    \sqrt{\frac{2}{T}}\sin\left[ \left(n - \frac{N - 1}{2}\right)\omega\right], &  n > \frac{N+1}{2}
    \end{matrix}
\right.
\end{align}
$$

$$
\int_{-T/2}^{T/2}f_n(\omega)f_m(\omega) = \delta_{m, n}
$$

# Harmonic trap

```python
w0 = 1
N = 201
w = .1
levels = 5


def V(x, w0=w0):
    return (w0*x)**2 / 2


p1d = Potential1D(V=V, N=N, w=w, verbose=1, points=[])
vals = p1d.get_eigenvals(levels=levels)
wf = p1d.get_eigenwf(xs=np.linspace(-10, 10, 1000))
print('bound states eigenergy is: ', vals)

for j in range(levels):
    plt.plot(p1d.xs, wf[j] + vals[j],
             label=r'$E_n='+f'{vals[j]:.2f}'+r'$')

plt.plot(p1d.xs, V(p1d.xs))
plt.xlabel(r'$x$')
plt.ylabel(r'Eigen wave function $\psi_n(x)$ for $n$th bound state')
plt.xlim(-5, 5)
plt.ylim(-.5, 6.5)
plt.legend()
plt.show()
```

# Finite well potential

```python
V0 = 50
a = 1
N = 201
w = 1
levels = 4


def V(x, V0=V0, a=a):
    return np.sign(np.abs(x) - a/2) * (V0/2) + V0/2


p1d = Potential1D(V=V, N=N, w=w, verbose=1, points=[-a/2, a/2])
vals = p1d.get_eigenvals(levels=levels)
wf = p1d.get_eigenwf(xs=np.linspace(-a, a, 1000))
print('bound states eigenergy is: ', vals)

for j in range(levels):
    plt.plot(p1d.xs, wf[j] + vals[j],
             label=r'$E_n='+f'{vals[j]:.2f}'+r'$')

plt.plot(p1d.xs, V(p1d.xs))
plt.xlabel(r'$x$')
plt.ylabel(r'Eigen wave function $\psi_n(x)$ for $n$th bound state')
plt.xlim(-a, a)
plt.ylim(-5, 55)
plt.legend()
plt.show()
```

# Double well

```python
V0 = 100
a = 1
N = 2001
w = 1
levels = 4


def V(x, V0=V0, a=a):
    V1 = np.sign(np.abs(x) - a/2) * (V0/2) + V0/2
    V2 = np.sign(a/8 - np.abs(x)) * (V0/4) + V0/4
    return V1+V2


p1d = Potential1D(V=V, N=N, w=w, verbose=1, points=[-a/2, a/2])
vals = p1d.get_eigenvals(levels=levels)
wf = p1d.get_eigenwf(xs=np.linspace(-2*a, 2*a, 1000))
print('bound states eigenergy is: ', vals)

for j in range(levels):
    plt.plot(p1d.xs, wf[j] + vals[j],
             label=r'$E_n='+f'{vals[j]:.2f}'+r'$')

plt.plot(p1d.xs, V(p1d.xs))
plt.xlabel(r'$x$')
plt.ylabel(r'Eigen wave function $\psi_n(x)$ for $n$th bound state')
plt.xlim(-a, a)
plt.ylim(-5, 105)
plt.legend()
plt.show()
```
